```python
import random

bilgisayar_sayisi = random.randint(1, 10)

print("1 ile 10 arasında bir sayı tuttum. Bakalım tahmin edebilecek misin?")

deneme_sayisi = 0

while True:
    deneme_sayisi += 1
    try:
        tahmin = int(input("Tahminini gir: "))
        if not (1 <= tahmin <= 10):
            print("Lütfen 1 ile 10 arasında bir sayı girin.")
            continue
    except ValueError:
        print("Geçersiz giriş. Lütfen bir sayı girin.")
        continue

    if tahmin == bilgisayar_sayisi:
        print(f"Tebrikler, doğru tahmin! Kutlarım! {deneme_sayisi} denemede bildin.")
        break
    elif tahmin < bilgisayar_sayisi:
        print("Daha büyük bir sayı dene.")
    else:
        print("Daha küçük bir sayı dene.")
```
